```python
#import all library needed
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

#load test image
img = cv.imread('sample_paper.png')
```

```python
#gaussian filter
gaussian_filter = cv.GaussianBlur(img, (3,3), 85)
cv.imwrite('gaussian.jpg', gaussian_filter)
```

```python
#median filter
median_filter = cv.medianBlur(img, 3)
cv.imwrite('median.jpg', median_filter)
```

```python
#average filter
avg_filter = cv.blur(img, (3,3))
cv.imwrite('avg.jpg', avg_filter)
```

```python
#add on: bilateral filter
bliateral_filter = cv.bilateralFilter(img,20,60,60)
cv.imwrite('bilateral.jpg', bliateral_filter)
```

```python
#try implement rectangle
start_point = (80,240)
end_point = (105,265)
color = (255,0,0)
thick = 2
image = cv.rectangle(img, start_point, end_point,color,thick)
plt.imshow(image)
plt.show()
```
